```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-regioni/dpc-covid19-ita-regioni-latest.csv")
```

```python
df.head(5)
```

```python
df_plot = df[["denominazione_regione", "ricoverati_con_sintomi"]].sort_values(by="ricoverati_con_sintomi", ascending=False)
```

```python
df_plot.plot(kind="bar", x="denominazione_regione", figsize=(15,10))
```
