```python
import pandas as pd 
```

```python
data = { 
    'Region': [ 
        'East', 'East', 'East', 'North', 'North', 'North', 'South', 'South',
        'South', 'West', 'West', 'West'],
    'Types': [ 
        'Children Clothing', 'Men Clothing', 'Women Clothing',
        'Children Clothing', 'Men Clothing', 'Women Clothing',
        'Children Clothing', 'Men Clothing', 'Women Clothing',
        'Children Clothing', 'Men Clothing', 'Women Clothing'],

    'SumSales': [45849, 51685, 70229, 37306, 39975, 61419,
                18570, 18542, 22203, 20182, 19077, 22217],

    'NumSales': [113, 182, 176, 85, 89, 142, 45, 39, 53, 42, 41, 53],
}
```

```python
len(data['Region']), len(data['Types']), len(data['SumSales']), len(data['NumSales'])
```

```python
df = pd.DataFrame.from_dict(data)
df['PercSales'] = df['SumSales'] / df['SumSales'].sum()
```

```python
df.head()
```

## Apply Currency and Percent Formatting

```python
df.style.format({'SumSales': '€ {0:,.0f}'})
```

```python
df.style.format({'SumSales': '€ {0:,.0f}', 'PercSales': '{:.0%}'})
```

## Apply Padding

```python
col_formatters = {'SumSales': '€ {0:,.0f}', 'PercSales': '{:.0%}', 'NumSales': '{:0>3d}'}
df.style.format(col_formatters)
```

## Highlighting Values in Pandas using Color

```python
# Find Maximum Values in DF
numeric_columns = ['SumSales', 'NumSales', 'PercSales']
df.style.highlight_max(subset=numeric_columns, color='lightgreen')
```

```python
df.style.highlight_min(subset=numeric_columns, color='red')
```

```python
df.style\
    .highlight_min(subset=numeric_columns, color='red')\
    .highlight_max(subset=numeric_columns, color='darkgreen')\
    .format(col_formatters)
```

## Highlight Ranges

```python
df.style.highlight_between(left=15000, right=40000, subset='SumSales', color='darkgreen')
```

## Highlight Quantiles

```python
df.style.highlight_quantile(q_left=0.25, q_right=0.75, subset='NumSales', color='darkgreen')
```

## Defining Custon Colors

```python
def add_color(x):
    if x < 50:
        color = 'red'
    elif x < 100:
        color = 'orange'
    else: 
        color = 'darkgreen'
    return f'background: {color}'

df.style.applymap(add_color, subset='NumSales')
```

## Using ColorMaps

```python
df.style.background_gradient(cmap='PiYG', subset=numeric_columns)
```

```python
import seaborn as sns
```

```python
colormap = sns.light_palette('green', as_cmap=True)
df.style.background_gradient(cmap=colormap, subset=numeric_columns)
```
