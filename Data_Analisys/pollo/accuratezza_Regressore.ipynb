```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import seaborn as sns
```

```python
y = np.random.random(size=100) * 10
errors = 2 * (np.random.random(size=100)) - 1
```

```python
p = y + errors
```

```python
mse = mean_squared_error(y, p)
mae = mean_absolute_error(y, p)
```

```python
res = y - p
sns.scatterplot(x=y, y=res)
```

Essendo i punti sparsi, senza un pattern sistematico, poco potremo fare per migliorare la situazione. Diversamente nel caso seguente...

```python
y = np.random.random(size=100) * 10
errors = y/2 + (2 * (np.random.random(size=100)) - 1)
p = y + errors
mse = mean_squared_error(y, p)
mae = mean_absolute_error(y, p)
res = y - p
sns.scatterplot(x=y, y=res)
```

La ***sistematicità*** degli errori si distribuiscono su di una retta. Possiamo intervenire per correggere il modello aggiungendo o modificando dati.
