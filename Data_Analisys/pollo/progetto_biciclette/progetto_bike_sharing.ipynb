```python
import pandas as pd
import seaborn as sns # libreria grafici as uso statistico
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Bike-Sharing-Dataset/hour.csv")
```

```python
len(df.index)
```

```python
df.sample(3)
```

```python
df.describe()
```

```python
sns.barplot(data=df, x="hr", y="cnt") # distribuzione prenotazioni durante la giornata
```

```python
df.columns
```

```python
for c in ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']:
          sns.barplot(data=df, x=c, y='cnt')
          plt.show()
```

```python
for c in ['temp', 'atemp', 'hum', 'windspeed']: # essendo variabili numeriche scatter è utile
    sns.scatterplot(data=df, x=c, y='cnt')
    plt.show()
```

```python
sns.jointplot(data=df, x='temp', y='cnt', kind='kde', fill=True)
```

```python
sns.pairplot(data=df[['temp', 'hum', 'windspeed', 'cnt']])
```

## Visualizzare i dati in 4 dimensioni

```python
sns.barplot(data=df, x='hr', y='cnt', hue='workingday')
# 0 weekend 1 lavorativo
```

```python
sns.relplot(data=df, x='hr', y='cnt', hue='season', col='workingday', kind='line')
```

```python
sns.relplot(data=df, x='temp', y='hum', hue='workingday', kind='scatter', col='yr', row='weathersit', size='cnt')
```

```python
sns.boxplot(data=df, x='season', y='cnt', hue='holiday')
```

```python
sns.violinplot(data=df, x='season', y='cnt', hue='holiday')
```
