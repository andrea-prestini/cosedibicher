```python
import pandas as pd
import seaborn as sns # libreria grafici as uso statistico
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Bike-Sharing-Dataset/hour.csv")
```

```python
df.sample(3)
```

```python
df.describe()
```

```python
sns.barplot(data=df, x="hr", y="cnt") # distribuzione prenotazioni durante la giornata
```

```python
df.columns
```

```python
for c in ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']:
          sns.barplot(data=df, x=c, y='cnt')
          plt.show()
```

```python
for c in ['temp', 'atemp', 'hum', 'windspeed']: # essendo variabili numeriche scatter  utile
    sns.scatterplot(data=df, x=c, y='cnt')
    plt.show()
```

```python
sns.jointplot(data=df, x='temp', y='cnt', kind='kde', shade=True)
```
