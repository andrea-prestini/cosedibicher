```python
try:
    t.loc["BS"] # non presente nome indice (None) viene usato un numero messo da Pandas
except:
    print("valore non presente!")
```

Nello $\color{red}{slicing}$ ricordiamo che:
    * loc include il riferimento esterno maggiore [1:4] 4 è compreso!
    * iloc NON include il riferimento esterno maggiore [1:4] 4 NON è compreso!

La sintassi più leggibile è df.loc\[righe, colonne\] da preferire sempre: nel caso in cui vogliamo tutte le colonne useremo la sintassi
~~~
df.loc[2:3, :]
~~~
